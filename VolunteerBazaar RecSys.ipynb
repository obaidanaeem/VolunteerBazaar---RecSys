## Generate Recommendations

```python
import numpy as np
from fuzzywuzzy import process
```

## Load Ground_Truth_Data.csv

```python
# Load and process Ground_Truth_Data.csv
gt_df = pd.read_csv("Ground_Truth_Data_3.csv")
# Normalize event names and volunteer names
gt_df['Event_Name'] = gt_df['Event_Name'].str.strip().str.lower()
# Assuming Volunteers_Participated is stored as a string (e.g., "vol1, vol2, vol3")
def process_volunteers(x):
    return [v.strip().lower() for v in x.split(",")] if isinstance(x, str) else []
gt_df['Volunteers_Participated'] = gt_df['Volunteers_Participated'].apply(process_volunteers)

# Build dictionary: {event_name: list_of_volunteer_names}
ground_truth_dict = dict(zip(gt_df['Event_Name'], gt_df['Volunteers_Participated']))
```

## Recommendation Function

```python
def get_volunteer_recommendations_2(df1, df1_encoded, df2, df2_encoded, similarity_df, ground_truth_dict=None, n_recommendations=5, weights=None):
    """
    Generates volunteer recommendations for a given event.
    
    Incorporates:
      - Content-based matching: location, skills/event types, rating.
      - Precomputed similarity scores.
      - A participation component (using normalized 'Activities_Volunteered' from df2).
      - If ground_truth_dict (a dict mapping event names to lists of volunteer names) is provided and contains the event,
        restricts the recommendation pool to that set of volunteers.
    
    Cases handled:
      - Case 1 (Few Volunteers): For very low participation counts, a moderate default is applied.
      - Case 2 (No Past Participation): The participation component will be near zero, so other features dominate.
      - Case 3 (Cold Start): New volunteers or events rely on content-based features and similarity scores.
      - Additionally, if an event has ground truth data for Volunteers_Participated, the recommendations are chosen only from that pool.
    """
    # Extend weights to include participation
    if weights is None:
        weights = {
            'location': 0.3,
            'skills_event_types': 0.3,
            'rating': 0.2,
            'similarity': 0.3,
            'participation': 0.3
        }
    try:
        # Get event name input from the console
        event_name = input("Enter Event Name: ").strip().lower()
        
        # Validate event existence in df1 (Volunteer_Activities.csv)
        if event_name not in df1['Title'].str.strip().str.lower().values:
            suggestions = process.extract(event_name, df1['Title'].str.lower().values, limit=3)
            print(f"Event '{event_name}' not found. Did you mean:")
            for suggestion in suggestions:
                print(f"- {suggestion[0]} (Confidence: {suggestion[1]}%)")
            return None
        
        # Get event index using normalized title comparison
        event_idx = df1[df1['Title'].str.strip().str.lower() == event_name].index[0]
        event_features = df1_encoded.loc[event_idx]
        
        # Restrict volunteer pool if ground_truth_dict is provided and the event has ground truth data
        if ground_truth_dict is not None and event_name in ground_truth_dict:
            pool_names = [name.strip().lower() for name in ground_truth_dict[event_name]]
            # Create a boolean mask for volunteers in df2 whose names appear in the ground truth pool
            mask = df2['Name'].str.strip().str.lower().isin(pool_names)
            if mask.sum() > 0:
                df2_pool = df2[mask].reset_index(drop=True)
                df2_encoded_pool = df2_encoded[mask].reset_index(drop=True)
                print(f"Using ground truth volunteer pool for event '{event_name}'. Total volunteers in pool: {len(df2_pool)}")
            else:
                print(f"Ground truth volunteer pool for event '{event_name}' is empty after filtering. Using full pool.")
                df2_pool = df2.copy()
                df2_encoded_pool = df2_encoded.copy()
        else:
            df2_pool = df2.copy()
            df2_encoded_pool = df2_encoded.copy()
        
        # Define feature columns (excluding non-informative ones)
        feature_cols = [col for col in df1_encoded.columns if col not in ['City', 'Rating', 'Organization']]
        
        # ----- Content-Based Components -----
        # 1. Location matching
        location_scores = (df2_encoded_pool['City'] == event_features['City']).astype(float) * weights['location']
        
        # 2. Skills and Event Types matching
        matching_features = (df2_encoded_pool[feature_cols].values * event_features[feature_cols].values).sum(axis=1)
        total_event_features = event_features[feature_cols].sum()
        feature_scores = (matching_features / total_event_features) * weights['skills_event_types'] if total_event_features > 0 else 0
        
        # 3. Rating contribution
        rating_scores = (df2_encoded_pool['Rating'] / 5.0) * weights['rating']
        
        # Combine content-based scores
        relevance_scores = location_scores + feature_scores + rating_scores
        
        # ----- Participation Component -----
        # Use "Activities_Volunteered" (count) from Volunteers_Data.csv (df2_pool) as participation
        participation_raw = df2_pool['Activities Volunteered'].astype(float)
        max_participation = participation_raw.max() if participation_raw.max() > 0 else 1
        participation_scores = (participation_raw / max_participation)
        
        # Case 1: For very low participation, assign a moderate default value to avoid penalizing too harshly.
        participation_scores = participation_scores.apply(
            lambda x: 0.5 if x < (1 / max_participation) else x
        )
        participation_scores = participation_scores * weights['participation']
        
        # ----- Similarity Component -----
        # Note: similarity_df is computed over all volunteers in the full pool.
        # We need to select rows corresponding to the volunteer pool (based on the filtering above).
        # If the original similarity_df has the same ordering as df2, then after filtering, we must reindex similarity scores.
        # One approach is to use df2_pool's index to get the corresponding similarity scores.
        # Here, we assume that df2 and similarity_df are aligned. So we filter similarity_df accordingly.
        similarity_scores_full = similarity_df.iloc[:, event_idx].values
        # Create an index mapping from full df2 to the pool:
        pool_indices = df2_pool.index.values
        similarity_scores = similarity_scores_full[pool_indices]
        
        # ----- Final Score Combination -----
        final_scores = weights['similarity'] * similarity_scores + relevance_scores + participation_scores
        
        # Get top N volunteers based on final scores.
        top_indices = np.argsort(final_scores)[-n_recommendations:][::-1]
        recommendations = pd.DataFrame({
            'Volunteer Name': df2_pool.iloc[top_indices]['Name'].values,
            'Total Score': final_scores[top_indices].round(3)
        })
        
        return recommendations.sort_values('Total Score', ascending=False)
    
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None
```

## Evaluation Metrics

```python
import pandas as pd
import numpy as np
from fuzzywuzzy import process

def get_volunteer_recommendations_2(df1, df1_encoded, df2, df2_encoded, similarity_df, ground_truth_dict=None, n_recommendations=5, weights=None):
    """
    Generates volunteer recommendations for all events in ground_truth_dict.
    
    Incorporates:
      - Content-based matching: location, skills/event types, rating.
      - Precomputed similarity scores.
      - A participation component (using normalized 'Activities_Volunteered' from df2).
      - If ground_truth_dict (a dict mapping event names to lists of volunteer names) is provided and contains the event,
        restricts the recommendation pool to that set of volunteers.
    
    Cases handled:
      - Case 1 (Few Volunteers): For very low participation counts, a moderate default is applied.
      - Case 2 (No Past Participation): The participation component will be near zero, so other features dominate.
      - Case 3 (Cold Start): New volunteers or events rely on content-based features and similarity scores.
      - Additionally, if an event has ground truth data for Volunteers_Participated, the recommendations are chosen only from that pool.
    """
    # Extend weights to include participation
    if weights is None:
        weights = {
            'location': 0.3,
            'skills_event_types': 0.3,
            'rating': 0.2,
            'similarity': 0.3,
            'participation': 0.3
        }
    
    all_recommendations = []

    try:
        # Iterate over all events in ground_truth_dict
        for event_name in ground_truth_dict.keys():
            event_name = event_name.strip().lower()
            
            # Validate event existence in df1 (Volunteer_Activities.csv)
            if event_name not in df1['Title'].str.strip().str.lower().values:
                suggestions = process.extract(event_name, df1['Title'].str.lower().values, limit=3)
                print(f"Event '{event_name}' not found. Did you mean:")
                for suggestion in suggestions:
                    print(f"- {suggestion[0]} (Confidence: {suggestion[1]}%)")
                continue
            
            # Get event index using normalized title comparison
            event_idx = df1[df1['Title'].str.strip().str.lower() == event_name].index[0]
            event_features = df1_encoded.loc[event_idx]
            
            # Restrict volunteer pool if ground_truth_dict is provided and the event has ground truth data
            pool_names = [name.strip().lower() for name in ground_truth_dict[event_name]]
            mask = df2['Name'].str.strip().str.lower().isin(pool_names)
            df2_pool = df2[mask].reset_index(drop=True)
            df2_encoded_pool = df2_encoded[mask].reset_index(drop=True)

            if df2_pool.empty:
                print(f"Ground truth volunteer pool for event '{event_name}' is empty after filtering. Skipping this event.")
                continue
            
            print(f"Using ground truth volunteer pool for event '{event_name}'. Total volunteers in pool: {len(df2_pool)}")
            
            # Define feature columns (excluding non-informative ones)
            feature_cols = [col for col in df1_encoded.columns if col not in ['City', 'Rating', 'Organization']]
            
            # ----- Content-Based Components -----
            # 1. Location matching
            location_scores = (df2_encoded_pool['City'] == event_features['City']).astype(float) * weights['location']
            
            # 2. Skills and Event Types matching
            matching_features = (df2_encoded_pool[feature_cols].values * event_features[feature_cols].values).sum(axis=1)
            total_event_features = event_features[feature_cols].sum()
            feature_scores = (matching_features / total_event_features) * weights['skills_event_types'] if total_event_features > 0 else 0
            
            # 3. Rating contribution
            rating_scores = (df2_encoded_pool['Rating'] / 5.0) * weights['rating']
            
            # Combine content-based scores
            relevance_scores = location_scores + feature_scores + rating_scores
            
            # ----- Participation Component -----
            participation_raw = df2_pool['Activities Volunteered'].astype(float)
            max_participation = participation_raw.max() if participation_raw.max() > 0 else 1
            participation_scores = (participation_raw / max_participation)
            
            # Case 1: For very low participation, assign a moderate default value to avoid penalizing too harshly.
            participation_scores = participation_scores.apply(
                lambda x: 0.5 if x < (1 / max_participation) else x
            )
            participation_scores = participation_scores * weights['participation']
            
            # ----- Similarity Component -----
            similarity_scores_full = similarity_df.iloc[:, event_idx].values
            pool_indices = df2_pool.index.values
            
            # Ensure we only access valid indices
            if len(pool_indices) > 0:
                similarity_scores = similarity_scores_full[pool_indices]
            else:
                print(f"No valid indices found for similarity scores for event '{event_name}'. Skipping this event.")
                continue
            
            # ----- Final Score Combination -----
            final_scores = weights['similarity'] * similarity_scores + relevance_scores + participation_scores
            
            # Get top N volunteers based on final scores.
            top_indices = np.argsort(final_scores)[-n_recommendations:][::-1]
            recommendations = pd.DataFrame({
                'Volunteer Name': df2_pool.iloc[top_indices]['Name'].values,
                'Total Score': final_scores[top_indices].round(3),
                'Event Name': event_name
            })
            
            all_recommendations.append(recommendations)

        # Combine all recommendations into a single DataFrame
        final_recommendations = pd.concat(all_recommendations, ignore_index=True)
        final_recommendations.to_csv('Predicted_Recommendations.csv', index=False)
        print("Recommendations saved to 'Predicted_Recommendations.csv'.")

    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load the predicted recommendations
predicted_df = pd.read_csv('Predicted_Recommendations.csv')

# Print the columns of the predicted DataFrame for debugging
print("Predicted DataFrame Columns:", predicted_df.columns)

# Load the ground truth data from Ground_Truth_Data.csv
ground_truth_df = pd.read_csv('Ground_Truth_Data_3.csv')

# Print the columns of the ground truth DataFrame for debugging
print("Ground Truth DataFrame Columns:", ground_truth_df.columns)

# Convert the ground truth DataFrame into a format suitable for merging
ground_truth_data = []
for index, row in ground_truth_df.iterrows():
    event = row['Event_Name']
    # Check if Volunteers_Participated is not NaN and is a string
    if pd.notna(row['Volunteers_Participated']):
        volunteers = str(row['Volunteers_Participated']).split(',')  # Convert to string and split
        for volunteer in volunteers:
            ground_truth_data.append({'Event': event.strip(), 'Volunteer': volunteer.strip(), 'Actual': 1})  # Actual = 1 for ground truth

# Create a DataFrame from the ground truth data
ground_truth_df = pd.DataFrame(ground_truth_data)

# Rename columns in predicted_df to match ground_truth_df for merging
predicted_df.rename(columns={
    'Event Name': 'Event',
    'Volunteer Name': 'Volunteer'
}, inplace=True)

# Merge predicted recommendations with ground truth data
merged_df = pd.merge(predicted_df, ground_truth_df, on=['Event', 'Volunteer'], how='left')

# Fill NaN values in 'Actual' column with 0 (indicating not in ground truth)
merged_df['Actual'] = merged_df['Actual'].fillna(0)

# Calculate predicted values based on whether the volunteer is in the ground truth
merged_df['Predicted'] = (merged_df['Total Score'] > 0).astype(int)

# Calculate evaluation metrics
y_true = merged_df['Actual'].values  # Ground truth values
y_pred = merged_df['Predicted'].values  # Predicted values

# Calculate evaluation metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred)

# Print the results
print("Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```
